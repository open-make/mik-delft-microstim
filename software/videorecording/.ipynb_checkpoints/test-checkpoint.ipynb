```python
from pypylon import pylon
import cv2, time
import numpy as np
from queue import Queue
```

```python
def display_video(path):
    # Create a VideoCapture object and read from input file
    # It needs a video file or camera path
    cap = cv2.VideoCapture(path)

    # Read until video is completed
    while(cap.isOpened()):
        # Capture frame-by-frame
        ret, frame = cap.read()
        if ret == True:
            # Display the resulting frame
            cv2.imshow('Frame',frame)
    
            # Press Q on keyboard to  exit
            if cv2.waitKey(25) & 0xFF == ord('q'):
                break
        # Break the loop
        else: 
            break
    
    # When everything done, release the video capture object and close all the frames
    cap.release()
    cv2.destroyAllWindows()
    return

videoName = 'C:/Users/miksc/OneDrive/coding_projects/SNA-100100_0.mp4'
```

```python
# Set up the camera
camera = pylon.InstantCamera(pylon.TlFactory.GetInstance().CreateFirstDevice())
camera.Open()

# Configure your settings
fps = 150 # Pupil=100, Whisker=200
frame_count = 0
frame_total = 120 # Total number of frames that will be saved. Trigger will be given in the middle
frameQueue = Queue(maxsize = frame_total)

# Set camera parameters (e.g., frame rate, exposure time)
camera.AcquisitionFrameRateEnable.SetValue(True)
camera.AcquisitionFrameRate.SetValue(fps)
camera.ExposureAuto.SetValue("Off")
camera.ExposureTime.SetValue(10000)

# Trigger with GPIO input
camera.LineSelector ='Line4'
camera.LineMode = 'Output'

# Set up the video writer
fourcc = cv2.VideoWriter_fourcc(*'h264') # Whats difference between mp4v, x264 and h264?
out = cv2.VideoWriter("output.mp4", fourcc, float(fps), (camera.Width.Value, camera.Height.Value))

triggered = False

# Start continuous acquisition
camera.StartGrabbing(pylon.GrabStrategy_LatestImageOnly)
while camera.IsGrabbing() and frame_count < (frame_total/2):
    # Check if the queue is full
    if frameQueue.full():
        removedFrame = frameQueue.get()

    grab_result = camera.RetrieveResult(5000, pylon.TimeoutHandling_ThrowException)
    if grab_result.GrabSucceeded():
        frame = grab_result.Array
        frameQueue.put(frame)
   
        cv2.imshow("Preview", frame)

        if triggered:
            # Save the frame to the video file
            frame_count += 1

        if camera.LineStatus():
            print('TRIGGERED')
            triggered = True

    grab_result.Release()

    if cv2.waitKey(1) == 27:
        break

start = time.time()
frameQueue = np.array(frameQueue.queue)
for frame in frameQueue:
    out.write(frame)
end = time.time()

print(f'It took {end-start} seconds to write output file')

# Release resources
camera.StopGrabbing()
camera.Close()
out.release()
cv2.destroyAllWindows()
```

```python
''''Here I will try to make it into a function and the trigger is given by pressing T'''

def capture_behaviour(camera, fileName, fps, frame_total):
    '''docstring
    
        INPUT:
            camera
            fileName
            fps
            frame_total
            
        OUTPUT: 
            singleTrialbehaviour
    '''
    # Prepare fileName and set up the video writer
    fileName = fileName + '.mp4'
    frame_count = 0
    triggered = False
    frameQueue = Queue(maxsize = frame_total)

    # Start continuous acquisition
    while camera.IsGrabbing() and frame_count < (frame_total/2):
        # Check if the queue is full
        if frameQueue.full():
            removedFrame = frameQueue.get()

        grab_result = camera.RetrieveResult(5000, pylon.TimeoutHandling_ThrowException)
        if grab_result.GrabSucceeded():
            frame = grab_result.Array
            frameQueue.put(frame)
    
            cv2.imshow("Preview", frame)

            if triggered:
                # Save the frame to the video file
                frame_count += 1

            key = cv2.waitKey(1) & 0xFF
            if key == ord('t'):
                print('TRIGGERED')
                triggered = True

        grab_result.Release()

        if cv2.waitKey(1) == ord('q'):
            return

    frameQueue = np.array(frameQueue.queue)
    fourcc = cv2.VideoWriter_fourcc(*'h264') # Whats difference between mp4v, x264 and h264?
    out = cv2.VideoWriter(fileName, fourcc, float(fps), (camera.Width.Value, camera.Height.Value))
    for frame in frameQueue:
        out.write(frame)
    out.release()
    return

```

```python
# Set up the camera
camera = pylon.InstantCamera(pylon.TlFactory.GetInstance().CreateFirstDevice())
camera.Open()

# Configure your settings
fps = 30 # Pupil=100, Whisker=200
frame_count = 0
frame_total = 120 # Total number of frames that will be saved. Trigger will be given in the middle
frameQueue = Queue(maxsize = frame_total)

# Set camera parameters (e.g., frame rate, exposure time)
camera.AcquisitionFrameRateEnable.SetValue(True)
camera.AcquisitionFrameRate.SetValue(fps)
camera.ExposureAuto.SetValue("Off")
camera.ExposureTime.SetValue(10000)

fileName_blank = 'SNA-100100'
iTrial = 0

camera.StartGrabbing(pylon.GrabStrategy_LatestImageOnly)
while True:
    fileName = fileName_blank + '_' + str(iTrial)
    capture_behaviour(camera, fileName, fps, frame_total)
    iTrial += 1

    if cv2.waitKey(1) == ord('q'):
        break
    
# Release resources
camera.StopGrabbing()
camera.Close()
cv2.destroyAllWindows() 
```
